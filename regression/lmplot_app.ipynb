```python
import seaborn as sns; sns.set_theme(color_codes=True)
tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               markers=["o", "x"])
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               palette="Set1")
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               palette=dict(Yes="g", No="r"))
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

```python
sns.lmplot(x="size", y="total_bill", hue="day", col="day",
               data=tips, height=6, aspect=.4)
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", hue="day",
               data=tips, col_wrap=2, height=3)
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",
               data=tips, height=3)
```

```python
g = sns.lmplot(x="total_bill", y="tip", row="sex", col="time",
               data=tips, height=3)
g = (g.set_axis_labels("Total bill (US Dollars)", "Tip")
      .set(xlim=(0, 60), ylim=(0, 12),
           xticks=[10, 30, 50], yticks=[2, 6, 10])
      .fig.subplots_adjust(wspace=.02))
```
