```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.set(style="darkgrid")
sns.kdeplot(data=tips, x="total_bill")
```

```python
sns.kdeplot(data=tips, y="total_bill")
```

```python
sns.kdeplot(data=tips)
```

```python
sns.kdeplot(data=tips, x="total_bill", bw_adjust=.3,shade=True)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5,cut=0)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="stack")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="fill")
```

```python
sns.kdeplot(
    data=tips, x="total_bill", hue="time",
    cumulative=True, common_norm=False, common_grid=True,
)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="size")
```

```python
sns.kdeplot(
   data=tips, x="total_bill", hue="size",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,
)
```

```python
sns.kdeplot(data=tips, x="total_bill", y="tip")
```

```python
sns.kdeplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.kdeplot(data=tips, x="total_bill", y="tip", hue="sex", fill=True)
```

```python
sns.kdeplot(data=tips, x="total_bill", y="tip", hue="sex", levels=5,thresh=.2)
```

```python
sns.kdeplot(data=tips, x="total_bill", y="tip", fill=True, levels=100,thresh=0,cmap="mako")
```
