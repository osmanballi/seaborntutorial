```python
import seaborn as sns
sns.set(style="darkgrid")
tips = sns.load_dataset("tips")
sns.swarmplot(x=tips["total_bill"])
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="total_bill", y="day", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="smoker",
...                    data=tips, palette="Blues", dodge=True,edgecolor="red",linewidth=.25,marker="*",size=10,) #YlOrBr
```

```python
sns.swarmplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
sns.swarmplot(x="time", y="tip", data=tips, size=6)
```
