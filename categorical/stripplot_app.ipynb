```python
import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.stripplot(x=tips["total_bill"])
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,palette="rocket")
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=False)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.05)
```

```python
sns.stripplot(x="total_bill", y="day", data=tips,
...                    jitter=True)
```

```python
sns.stripplot(x="total_bill", y="day", data=tips,
...                    jitter=True, linewidth=1,)
```

```python
sns.stripplot(x="sex", y="total_bill", hue="day",
...                    data=tips, jitter=True,dodge=True)
```

```python
sns.stripplot(x="day", y="total_bill", hue="smoker",
...                    data=tips, jitter=True,
...                    palette="Set2", dodge=True)
```

```python
sns.stripplot(x="day", y="tip", data=tips,
...                    order=["Fri", "Sun"])
```

```python
sns.stripplot("day", "total_bill", "smoker", data=tips,
...                    palette="Set2", size=20, marker="o",linewidth=1,
...                    edgecolor="gray", alpha=.75)
```
