```python
import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.violinplot(x=tips["total_bill"])
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
...                     data=tips, palette="YlOrBr")
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
...                     data=tips, palette="muted", split=True)
```

```python
sns.violinplot(x="time", y="tip", data=tips,
...                     order=["Dinner", "Lunch"])
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",
...                     data=tips, palette="Set3", split=True,
...                     scale="area",inner="quartile") #inner=point, box, stick
```

```python
sns.set(style="darkgrid")
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2", split=True,scale="count")
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips,dodge=True,)
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set2", split=True,
                    scale="count", inner="stick", scale_hue=False)
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set2", split=True,
                    scale="count", inner="stick",
                    scale_hue=False, bw=.2)
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
ax = sns.violinplot(x="day", y="total_bill", hue="weekend",
                    data=tips, dodge=False)
```
