```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
print(tips.columns)
```

```python
print(tips.head(5))
```

```python
tips.isnull().sum().sort_values(ascending=False)
```

```python
tips.smoker.value_counts()
```

```python
sns.barplot(data = tips, x = "day", y = "total_bill")
```

```python
sns.barplot(x="tip", y="day", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.barplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
sns.barplot(x="day", y="tip", data=tips, capsize=.2,edgecolor=(0,0,0),
                  linewidth=2)
```

```python
sns.barplot(x="size", y="total_bill", data=tips, color="salmon", saturation=.5)
sns.despine(top=True, right=True)
```

```python
sns.barplot("size", y="total_bill", data=tips, palette="Blues_d")
```

```python
sns.barplot("size", y="total_bill", data=tips, palette="vlag")
```

```python
sns.barplot("size", y="total_bill", data=tips, palette="rocket")
```

```python
sns.barplot(x="day", y="total_bill", data=tips, linewidth=2.5, facecolor=(1, 1, 1, 0), errcolor=".2", edgecolor=".2")
```

```python
sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="bar",
                height=4, aspect=.7)
```

```python
sns.countplot(x='sex',data=tips)
```
