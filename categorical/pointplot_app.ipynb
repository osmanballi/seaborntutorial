```python
import seaborn as sns
sns.set(style="darkgrid")
tips = sns.load_dataset("tips")
sns.pointplot(x="time", y="total_bill", data=tips)
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips)
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips, dodge=True)
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips,
                   markers=["o", "x"],
                   linestyles=["-", "--"])
```

```python
sns.pointplot(x="tip", y="day", data=tips)
```

```python
sns.pointplot(x="tip", y="day", data=tips, join=False)
```

```python
sns.pointplot(x="time", y="total_bill", data=tips,
                   color="#bb3f3f")
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips, palette="Set2")
```

```python
sns.pointplot(x="time", y="tip", data=tips,
                   order=["Dinner", "Lunch"])
```

```python
from numpy import median
sns.pointplot(x="day", y="tip", data=tips, estimator=median)
```

```python
sns.pointplot(x="day", y="tip", data=tips, ci="sd")
```

```python
sns.pointplot(x="day", y="tip", data=tips, capsize=.2,palette="rocket")
```
